### 4. 문자열

```python


# 이스케이프 문자
# '\ : 작은 따옴표 , "\ : 큰따옴표, \n : 줄바꿈, \t : tab
print("\"안녕\" 하세요")

# 여러 줄 문자열
print("""동해물과 백두산이 마르고 닳도록
하느님이 보우하사
우리나라 만세""")

# 문자열 연산자 
# 문자열 연결 연산자 '+'
print("문자열" + "문자열") 
print("문자열" +1) # TypeError , 문자열+숫자은 타입에러 발생

# 문자열 반복 연산자 '*'
print("안녕하세요" *3)

# 문자 선택 연산자, 인덱싱 []
print("안녕하세요"[0]) # 안
print("안녕하세요"[-1]) # 음수 입력하면 뒤에서 인덱싱

# 문자열 범위 선택자, 슬라이싱 [:]
# 파이썬 문자열 슬라이싱은 마지막 숫자를 포함하지 않음
print("안녕하세요"[1:4]) # 녕하세 | 안[0] 녕[1] 하[2] 세[3] 요[4]
# 대괄호 안에 넣는 숫자 둘중 하나 생략 가능 (e.g. [1:], [:3])
# 변수[start:stop:step] 
    # 슬라이싱 시작 인덱스 start, 종료 인덱스 stop, 슬라이싱 간격 step
print("안녕하세요"[::-1]) # 요세하녕안
print("안녕하세요"[::2]) # 안하요
    

# 문자열 길이 구하기 (=문자의 개수, 문자 길이)
print(len("안녕하세요")) # 5

```

### 5. 숫자

```python
# integer 정수형 : 소수점이 없는 숫자
# floating 실수형 : 소수점이 있는 숫자

# 숫자 연산자
# +, -, *, /, //(나머지 버리기, 정수 나누기 연산자), %(나머지), ** 제곱 연산자

print("5 + 7 =", 5 + 7)
print("5 - 7 =", 5 - 7)
print("5 / 7 =", 5 / 7)
print("5 * 7 =", 5 * 7)
print("5 // 7 =", 5 // 7)
print("5 % 7 =", 5 % 7)
print("5 ** 7 =", 5 ** 7)

# 연산자의 우선순위
# [곱셈&나눗셈] > [덧셈&뺄셈] 우선순위 가지는 연산자는 왼->오 계산 
# 연산자 우선순위가 헷갈리면 괄호를 사용.
print(5 + (3 * 2)) #11

# TypeError 예외 : 문자열 + 숫자 다른 타입으로 연산할 경우 발생

# 문자열 연산자 우선 순위 *가 우선
print("안녕" + "하세요" * 3) # 안녕하세요하세요하세요
print("안녕" + ("하세요" * 3)) # 안녕하세요하세요하세요
print("안녕하세요" * 3) # 안녕하세요안녕하세요안녕하세요

# 진수 변환 함수 
# 10 진수 -> 2, 8, 16 진수로 변환
bin()  # 2
oct() # 8
hex() # 16

# int 내장함수로 10진수 변환 가능
b = int('0b11010', 2) # 해당 값과 진수

# 정수 자료형 메서드 bit_count
x = 4
x.bit_count() # 출력: 1 , 4의 이진수 '100' 

# integer interning [-5 ~ 256]
# 해당 범위 안의 정수 값은 여러 변수여도 동일한 주소값(메모리할당)을 가짐 예외 case..


# float 값으로 연산하면 모두 실수로 결과 값이 나옴
# integer 값으로 '/' 연산하면 실수 결과 값이 나오는 경우가 있음

# float 특수값
x = float("inf") # 양의 무한대
y = float("-inf") # 음의 무한대

float("inf") > 10 # Ture

# 실수의 부동 소수점 문제
    # 컴퓨터 연산이 늘 정확하지 않아서 이를 고려하여 코딩한다. 즉 10진수 숫자를 표현하는데 완벽하지 않다. 이러한 점을 유념해서 코딩해야함
# 부동소수점 연산 오류를 방지하기 위해서 decimal 함수를 사용
num1 = Decimal('0.1')
num2 = Decimal('0.2')
result = num1 + num2 # 0.3

# Decial은 표준 라이브러리 모듈로 기본 내장 함수가 아니므로 임포트 필요.
from decimal import Decimal

# decimal 모듈 전체를 임포트
import decimal

# decimal 모듈에서 Decimal 클래스만 임포트
from decimal import Decimal

# 실수자료형 [4.나아가기], 복소수 나중에 더 살펴봐야함 ->이해 안됨 =_=

# 복합 대입 연산자





```
